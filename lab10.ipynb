```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
minutes = np.array([10, 20, 30, 40, 50, 60, 70]).reshape(-1, 1)
calories = np.array([50, 100, 150, 190, 240, 280, 330])

# Linear Regression model
model = LinearRegression()
model.fit(minutes, calories)

# Predict calories burned for 45 minutes
predicted_45 = model.predict([[45]])
print(f"Predicted calories burned for 45 minutes: {predicted_45[0]:.2f} calories")

# Visualization
plt.scatter(minutes, calories, color='blue', label='Actual Data')
plt.plot(minutes, model.predict(minutes), color='red', label='Regression Line')
plt.scatter(45, predicted_45, color='green', marker='x', s=100, label='Prediction (45 mins)')
plt.xlabel('Minutes of Exercise')
plt.ylabel('Calories Burned')
plt.title('Simple Linear Regression: Calories vs Minutes of Exercise')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
# Features: [RAM, Storage, Battery]
X = np.array([
    [4, 64, 3000],
    [6, 128, 3500],
    [8, 128, 4000],
    [6, 64, 3300],
    [8, 256, 4500],
    [12, 256, 5000]
])

# Target variable: Price
y = np.array([250, 320, 400, 300, 450, 520])

# Build and train the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a new phone: 8 GB RAM, 128 GB storage, 4200 mAh battery
new_phone = np.array([[8, 128, 4200]])
predicted_price = model.predict(new_phone)
print(f"Predicted price: ${predicted_price[0]:.2f}")

# Analyze feature impact
coefficients = model.coef_
intercept = model.intercept_

print("\nFeature Impact on Price:")
print(f"RAM (GB): {coefficients[0]:.2f} $ per GB")
print(f"Storage (GB): {coefficients[1]:.2f} $ per GB")
print(f"Battery (mAh): {coefficients[2]:.4f} $ per mAh")
print(f"Intercept (Base Price): {intercept:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
hours = np.array([6, 8, 10, 12, 14, 16, 18, 20]).reshape(-1, 1)
temps = np.array([15, 17, 20, 25, 28, 26, 22, 18])

# Polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(hours)

# Train model
model = LinearRegression()
model.fit(X_poly, temps)

# Predict temperature at 13:00 (1 PM)
hour_to_predict = np.array([[13]])
hour_poly = poly.transform(hour_to_predict)
predicted_temp = model.predict(hour_poly)
print(f"Predicted temperature at 13:00: {predicted_temp[0]:.2f}°C")

# Plotting
hours_range = np.linspace(6, 20, 100).reshape(-1, 1)
hours_range_poly = poly.transform(hours_range)
temp_preds = model.predict(hours_range_poly)

plt.scatter(hours, temps, color='blue', label='Actual Data')
plt.plot(hours_range, temp_preds, color='red', label='Polynomial Curve')
plt.scatter(13, predicted_temp, color='green', marker='x', s=100, label='Prediction (13:00)')
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (°C)')
plt.title('Polynomial Regression: Temperature vs Hour')
plt.legend()
plt.grid(True)
plt.show()
```
