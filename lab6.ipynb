```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=100
sigma=5
#Generate a normal distribution
data=np.random.normal(mu,sigma,10000)
#Calculate the PDF
x=np.linspace(mu-3*sigma,mu+3*sigma,10000)
pdf=norm.pdf(x,mu,sigma)
plt.figure(figsize=(10,6))
plt.hist(data,bins=30,density=True,alpha=0.6,color='#1f77b4',edgecolor='blue')
plt.plot(x,pdf,'r-',lw=2)
plt.title('Normal Distribution: $\mu$=100, $\sigma$=5')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set parameters
mean = 60
std_dev = 12
num_samples = 1000

# Generate random data from a normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Create a histogram to visualize the data
plt.figure(figsize=(10, 6))

# Plot the histogram with density=True to normalize the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label="Histogram")

# Plot the Probability Density Function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'k', linewidth=2, label="PDF")

# Add labels and title
plt.title('Normal Distribution Histogram and PDF')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set parameters
mean = 30
std_dev = 5

# Create an array of x values (calorie intakes)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the Probability Density Function (PDF) for the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# I. Probability of greater than or equal to 19 calories
prob_greater_than_19 = 1 - norm.cdf(19, mean, std_dev)

# II. Probability of less than or equal to 19 calories
prob_less_than_19 = norm.cdf(19, mean, std_dev)

# III. Probability of between 19 and 35 calories
prob_between_19_and_35 = norm.cdf(35, mean, std_dev) - norm.cdf(19, mean, std_dev)

# Print the calculated probabilities
print(f"Probability of greater than or equal to 19 calories: {prob_greater_than_19:.4f}")
print(f"Probability of less than or equal to 19 calories: {prob_less_than_19:.4f}")
print(f"Probability of between 19 and 35 calories: {prob_between_19_and_35:.4f}")

# Visualize the distribution and the areas of interest
plt.figure(figsize=(10, 6))

# Plot the normal distribution PDF
plt.plot(x, pdf, 'b-', label="Normal Distribution")

# Fill the area for "greater than or equal to 19 calories"
plt.fill_between(x, pdf, where=(x >= 19), color='red', alpha=0.3, label=">= 19 calories")

# Fill the area for "less than or equal to 19 calories"
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.3, label="<= 19 calories")

# Fill the area for "between 19 and 35 calories"
plt.fill_between(x, pdf, where=((x >= 19) & (x <= 35)), color='yellow', alpha=0.5, label="19 <= x <= 35")

# Add labels and title
plt.title('Probability Distribution of Daily Calorie Intake')
plt.xlabel('Daily Calorie Intake (calories)')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
mean = 30
std_dev = 5

# Create an array of x values (calorie intakes)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the Probability Density Function (PDF) manually
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# I. Probability of greater than or equal to 19 calories
z_19 = (19 - mean) / std_dev
prob_greater_than_19 = 1 - (0.5 * (1 + np.math.erf(z_19 / np.sqrt(2))))

# II. Probability of less than or equal to 19 calories
prob_less_than_19 = 0.5 * (1 + np.math.erf(z_19 / np.sqrt(2)))

# III. Probability of between 19 and 35 calories
z_35 = (35 - mean) / std_dev
prob_between_19_and_35 = (0.5 * (1 + np.math.erf(z_35 / np.sqrt(2)))) - prob_less_than_19

# Print the calculated probabilities
print(f"Probability of greater than or equal to 19 calories: {prob_greater_than_19:.4f}")
print(f"Probability of less than or equal to 19 calories: {prob_less_than_19:.4f}")
print(f"Probability of between 19 and 35 calories: {prob_between_19_and_35:.4f}")

# Visualize the distribution and the areas of interest
plt.figure(figsize=(10, 6))

# Plot the normal distribution PDF
plt.plot(x, pdf, 'b-', label="Normal Distribution")

# Fill the area for "greater than or equal to 19 calories"
plt.fill_between(x, pdf, where=(x >= 19), color='red', alpha=0.3, label=">= 19 calories")

# Fill the area for "less than or equal to 19 calories"
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.3, label="<= 19 calories")

# Fill the area for "between 19 and 35 calories"
plt.fill_between(x, pdf, where=((x >= 19) & (x <= 35)), color='yellow', alpha=0.5, label="19 <= x <= 35")

# Add labels and title
plt.title('Probability Distribution of Daily Calorie Intake')
plt.xlabel('Daily Calorie Intake (calories)')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
